```python
style_dir = os.listdir("素材風格(主圖)")
style_dir.remove(".DS_Store")
mutual_df = pd.DataFrame(columns=['style_dirs','mutual_pics'])

for i in style_dir:
    first = os.listdir("素材風格(主圖)/"+i)
    no = style_dir.index(i) + 1
    second_list = style_dir[no:]
    for j in second_list:
        second = os.listdir("素材風格(主圖)/"+j)
        mutual = [x for x in first for y in second if x == y]
        dirs = str(i)+" 與 "+str(j)
        new_row = {'style_dirs': dirs, 'mutual_pics': len(mutual)}
        mutual_df.loc[len(mutual_df)] =  new_row
    

```

```python
columns = ['E1_1','E1_2','E1_3','E1_4','E1_5','E1_6','E1_7','E1_8','E1_9','E1_10']
index = ['E1_1','E1_2','E1_3','E1_4','E1_5','E1_6','E1_7','E1_8','E1_9','E1_10']
m_df = pd.DataFrame(columns=columns,index=index)
m_df
```

```python
s_df = pd.read_excel("P1圖對應素材風格(主圖).xlsx")
c_list = m_df.columns.to_list()
i_list = m_df.index.to_list()
for column in tqdm(c_list):
    first = s_df[s_df['TagCode']== column].主要系列編號
    c = c_list.index(column)
    for index in i_list:
        second = s_df[s_df['TagCode']== index].主要系列編號
        mutual = [x for x in first for y in second if x == y]
        i = i_list.index(index)
        m_df.iloc[i,c] = len(mutual)
```
