```python
style_dir = os.listdir("素材風格(主圖)")
style_dir.remove(".DS_Store")
mutual_df = pd.DataFrame(columns=['style_dirs','mutual_pics'])

for i in style_dir:
    first = os.listdir("素材風格(主圖)/"+i)
    no = style_dir.index(i) + 1
    second_list = style_dir[no:]
    for j in second_list:
        second = os.listdir("素材風格(主圖)/"+j)
        mutual = [x for x in first for y in second if x == y]
        dirs = str(i)+" 與 "+str(j)
        new_row = {'style_dirs': dirs, 'mutual_pics': len(mutual)}
        mutual_df.loc[len(mutual_df)] =  new_row
    

```

### 讀取csv的方法

```python
s_df = pd.read_excel("P1圖對應素材風格(主圖).xlsx")
s_df.info()
```

```python
s1 = s_df[s_df['TagCode']=="E1_1"]
s1_df = pd.DataFrame(s1.主要系列編號)
s1_df = s1_df.rename({'主要系列編號':'E1_1'},axis=1)

s2 = s_df[s_df['TagCode']=="E1_2"]
s2_df = pd.DataFrame(s2.主要系列編號)
s2_df = s2_df.rename({'主要系列編號':'E1_2'},axis=1)

s3 = s_df[s_df['TagCode']=="E1_3"]
s3_df = pd.DataFrame(s3.主要系列編號)
s3_df = s3_df.rename({'主要系列編號':'E1_3'},axis=1)

s4 = s_df[s_df['TagCode']=="E1_4"]
s4_df = pd.DataFrame(s4.主要系列編號)
s4_df = s4_df.rename({'主要系列編號':'E1_4'},axis=1)

s5 = s_df[s_df['TagCode']=="E1_5"]
s5_df = pd.DataFrame(s5.主要系列編號)
s5_df = s5_df.rename({'主要系列編號':'E1_5'},axis=1)

s6 = s_df[s_df['TagCode']=="E1_6"]
s6_df = pd.DataFrame(s6.主要系列編號)
s6_df = s6_df.rename({'主要系列編號':'E1_6'},axis=1)

s7 = s_df[s_df['TagCode']=="E1_7"]
s7_df = pd.DataFrame(s7.主要系列編號)
s7_df = s7_df.rename({'主要系列編號':'E1_7'},axis=1)

s8 = s_df[s_df['TagCode']=="E1_8"]
s8_df = pd.DataFrame(s8.主要系列編號)
s8_df = s8_df.rename({'主要系列編號':'E1_8'},axis=1)

s9 = s_df[s_df['TagCode']=="E1_9"]
s9_df = pd.DataFrame(s9.主要系列編號)
s9_df = s9_df.rename({'主要系列編號':'E1_9'},axis=1)

s10 = s_df[s_df['TagCode']=="E1_10"]
s10_df = pd.DataFrame(s10.主要系列編號)
s10_df = s10_df.rename({'主要系列編號':'E1_10'},axis=1)

df = pd.concat([s1_df,s2_df,s3_df,s4_df,s5_df,s6_df,s7_df,s8_df,s9_df,s10_df], axis=1)
```

```python
columns = ['E1_1','E1_2','E1_3','E1_4','E1_5','E1_6','E1_7','E1_8','E1_9','E1_10']
index = ['E1_1','E1_2','E1_3','E1_4','E1_5','E1_6','E1_7','E1_8','E1_9','E1_10']
m_df = pd.DataFrame(columns=columns,index=index)
m_df
```

```python
s_df = pd.read_excel("P1圖對應素材風格(主圖).xlsx")
c_list = m_df.columns.to_list()
i_list = m_df.index.to_list()
for column in tqdm(c_list):
    first = s_df[s_df['TagCode']== column].主要系列編號
    c = c_list.index(column)
    for index in i_list:
        second = s_df[s_df['TagCode']== index].主要系列編號
        mutual = [x for x in first for y in second if x == y]
        i = i_list.index(index)
        m_df.iloc[i,c] = len(mutual)
```
