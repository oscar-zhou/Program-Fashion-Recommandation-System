```python
stylebook = pd.read_csv("stylebook1.csv")
stylebook
```

```python
main_tag_1 = pd.read_csv("style_table.csv",index_col=0)
main_tag_1
```

```python
outfit_info = main_tag_1[['No','StyleBookID','main_url','mainID']]
outfit_info['top1'] = 'null'
outfit_info['top2'] = 'null'
outfit_info['bottom'] = 'null'
outfit_info['accessory1'] = 'null'
outfit_info['accessory2'] = 'null'

outfit_info
```

```python

for i,item in outfit_info.iterrows():
    stylebookid = item[1]
    url = item[2]

    all_item = stylebook.loc[(stylebook['StyleBookID'] == stylebookid) & (stylebook['款主圖'] == url)]
    for j in all_item['貼標款']:
        if len(all_item['貼標款']) == 1:
            if j.startswith(('AB','FA','MA','QA','GA','EA','HB')):
                outfit_info['top1'][i] = j
            elif j.startswith(('BA','CA','DA','QA','KL','KG','SD','KS')):
                outfit_info['bottom'][i] = j
            elif j.startswith(('YA','ZQ','ZC')):
                outfit_info['accessory1'] = j
        elif len(all_item['貼標款']) == 2:
            if j.startswith(('AB','FA','MA','QA','GA','EA','HB')):
                if outfit_info['top1'][i] == "null":
                    outfit_info['top1'][i] = j
                else:
                    outfit_info['top2'][i] = j
            elif j.startswith(('BA','CA','DA','QA','KL','KG','SD','KS')):
                outfit_info['bottom'][i] = j
            elif j.startswith(('YA','ZQ','ZC')):
                if outfit_info['accessory1'][i] == "null":
                    outfit_info['accessory1'][i] = j
                else:
                    outfit_info['accessory2'][i] = j
        elif len(all_item['貼標款']) == 3:
            if j.startswith(('AB','FA','MA','QA','GA','EA','HB')):
                if outfit_info['top1'][i] == "null":
                    outfit_info['top1'][i] = j
                else:
                    outfit_info['top2'][i] = j
            elif j.startswith(('BA','CA','DA','QA','KL','KG','SD','KS')):
                outfit_info['bottom'][i] = j
            elif j.startswith(('YA','ZQ','ZC')):
                if outfit_info['accessory1'][i] == "null":
                    outfit_info['accessory1'][i] = j
                else:
                    outfit_info['accessory2'][i] = j
        elif len(all_item['貼標款']) == 4:
            if j.startswith(('AB','FA','MA','QA','GA','EA','HB')):
                if outfit_info['top1'][i] == "null":
                    outfit_info['top1'][i] = j
                else:
                    outfit_info['top2'][i] = j
            elif j.startswith(('BA','CA','DA','QA','KL','KG','SD','KS')):
                outfit_info['bottom'][i] = j
            elif j.startswith(('YA','ZQ','ZC')):
                if outfit_info['accessory1'][i] == "null":
                    outfit_info['accessory1'][i] = j
                else:
                    outfit_info['accessory2'][i] = j

        
```

### 圖片搭配跟款號搭配兩張表

圖片搭配table，款號跟搭配款如果重複，取圖片網址
網址後面有PIC的沒有什麼用

```python
img_outfit = pd.read_csv("圖片搭配.csv",index_col=0)
img_outfit
```

```python
img_outfit.loc[(img_outfit['款號']== "AA4614") & (img_outfit['搭配款']== "BA2198")]
```

```python
img_outfit.drop_duplicates(subset='圖片',inplace=True)
img_outfit
```

```python
img_outfit.loc[img_outfit['圖片'].str.contains('PIC|T9')]
```

```python
"PIC" in "https://image.obdesign.com.tw/catalog/products/AB10813/AB10813@PI-6.jpg" or "T9" in"https://image.obdesign.com.tw/catalog/products/AB10813/AB10813@T96.jpg"
```

```python
#圖片搭配除掉有一筆以上 主圖網址裡面有PIC的資料
for i,item in tqdm(img_outfit.iterrows()):
    main = item[0]
    sub = item[3]
    length = len(img_outfit.loc[(img_outfit['款號']== main) & (img_outfit['搭配款']== sub)])
    fil_df = img_outfit.loc[(img_outfit['款號']== main) & (img_outfit['搭配款']== sub)]
    if length > 1:
        for j,dup in fil_df.iterrows():
            url_name = str(dup[2])
            if len(fil_df) > 1:
                if "PIC" in url_name or "T9" in url_name:
                    fil_df.drop(fil_df.loc[fil_df['圖片'] == dup[2]].index,inplace=True)
                    img_outfit.drop(img_outfit.loc[img_outfit['圖片'] == dup[2]].index,inplace=True)
                elif "@1" in url_name or "@2" in url_name:
                    fil_df.drop(fil_df.loc[fil_df['圖片'] == dup[2]].index,inplace=True)
                    img_outfit.drop(img_outfit.loc[img_outfit['圖片'] == dup[2]].index,inplace=True)
```

```python
img_outfit.to_csv("圖片搭配(整理過).csv")
```
